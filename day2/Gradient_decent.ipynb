```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['xtick.major.size'] = 7
matplotlib.rcParams['xtick.labelsize'] = 'x-large'
matplotlib.rcParams['ytick.major.size'] = 7
matplotlib.rcParams['ytick.labelsize'] = 'x-large'
matplotlib.rcParams['xtick.top'] = False
matplotlib.rcParams['ytick.right'] = False
matplotlib.rcParams['ytick.direction'] = 'in'
matplotlib.rcParams['xtick.direction'] = 'in'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['figure.figsize'] = [7,7]
```

Consider the following function with one local minimum and one global minimum. 
We start from some intialization of the parameter and subsequently update the parameter with gradient decent. 

The goal is to eventually reach the global minimum of the toy function by gradient updates!

```
def f(x):
  '''toy function
  '''

  return -1* x**2 + .07* x**4. + 0.05* (x - 6)**2

def dfdx(x):
  '''the gradient of the toy funciton
  '''

  return -2*x + .28*x**3+.1*(x-6)  
```

Now let's visualize the function and it's gradient. 

We also show the starting point of the parmater ($x = -3.8$)

```
x = np.linspace(-4,4,100)
x0 = -3.8
plt.scatter(x0, f(x0), s = 100)
plt.plot(x, f(x))
plt.xlabel(r"$\theta$")
plt.ylabel(r"$J(\theta)$")
```

```
plt.scatter(x0, dfdx(x0), s = 100)
plt.plot(x, dfdx(x))
plt.xlabel(r"$\theta$")
plt.ylabel(r"$dJ/d\theta$")
```

```
eps_dict = {"high learning rate": 0.8, "medium learning rate": 0.4, "low learning rate": 0.05}

def GD_visualization():
  for key, values in eps_dict.items():  
    plt.figure()
    x0 = -3.9
    eps = values 
    plt.scatter(x0, f(x0), s = 400, alpha = 0.5, color = "C5")
    plt.text(x0, f(x0)+ 0.5, "starting point", fontsize = 15)
    plt.plot(x, f(x), lw = 3, color = "k")
    for t in range(0, 40):
      g = dfdx(x0)
      x0 = x0 - g * eps
      plt.scatter(x0, f(x0), s = 400, alpha = 0.5, color = "C5")
      if (t+1)%10 == 0:
        plt.text(x0, f(x0)+0.25+0.25*(t+1)/10, "iteration = "+str(t+1), fontsize = 15)
    plt.xlabel(r"$\theta$")
    plt.ylabel(r"$J(\theta)$")  
    plt.title(key , fontsize = 20)
```

```
GD_visualization()
```
